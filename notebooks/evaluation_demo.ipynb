# Model Evaluation Demo

This notebook demonstrates how to programmatically evaluate the model using `evaluate.py`.

```python
import pandas as pd
from evaluate import run_evaluation, load_config
```

```python
# Load configuration
config_path = "debug_config.yaml"
config = load_config(config_path)

# You can modify config programmatically here if needed
print("Model:", config['model']['name_or_path'])
print("Dataset:", config['data']['dataset_name'])
```

```python
# Run evaluation on a small subset (e.g., 5 samples) to test
metrics, samples = run_evaluation(config, max_samples=5)

print("Metrics:", metrics)
```

```python
# Display samples in a DataFrame
df = pd.DataFrame(samples)
pd.set_option('display.max_colwidth', None)
display(df[['completion', 'ground_truth', 'accuracy_score', 'format_score']])
```

```python
# Deep dive into a single sample
sample_idx = 0
print("Prompt:\n", df.iloc[sample_idx]['prompt'])
print("\nModel Completion:\n", df.iloc[sample_idx]['completion'])
print("\nGround Truth:\n", df.iloc[sample_idx]['ground_truth'])
```
